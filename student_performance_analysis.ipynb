```python
import pandas as pd

df = pd.read_csv("student_data.csv")
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df['G3'].hist()
plt.title("Final Grade Distribution")
plt.show()
```

## Insight 1: Distribution of Final Grades

The distribution of final grades (G3) appears to be approximately normal,
with most students scoring between 8 and 15.

This suggests that the majority of students perform at an average level,
with relatively few extreme high or low scores.

```python
import matplotlib.pyplot as plt

plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Final Grade")
plt.show()
```

## Insight 2: Study Time and Performance

There is a positive relationship between study time and final grades,
although the correlation is not very strong.

This suggests that while studying is important, other factors also influence performance.
